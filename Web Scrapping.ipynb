Extracting text from the website

```python
Title_1 = Soup.find("h3",class_="fs-20 lh-30 fw-600 label-color-5")
for title_1 in Title_1:
    print(title_1.text)

text_1 = Soup.find("p",class_="fs-16 fw-400 lh-24 label-color-1 card-text")
text_1

for para1 in text_1:
    print(para1.text)
```

```python
all_content = Soup.find_all("p")
for content in all_content:
    print(content.text)
```

Extracting Links from the web page

```python
all_headings = Soup.find_all("a")
for headings in all_headings:
    print(headings.get("href"))
all_headings
```

Getting the images 

```python
img = Soup.find_all("img")
for src in img:
    print(src.get("src"))
```
